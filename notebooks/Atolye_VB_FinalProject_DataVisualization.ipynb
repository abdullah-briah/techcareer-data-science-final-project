```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\home\Downloads\Atolye_VB-Bitirme-Projesi\Data Science Ödev dosyası 1\Ödev 2\50_Startups.csv")
```

```python
data.head()  # Display the first 5 rows of the dataset to preview the data structure
```

```python
plt.figure(figsize=(8,6))   # Create a blank figure of size 8x6 inches

# Output => <Figure size 800x600 with 0 Axes> 
# (This just means the figure was created but nothing was plotted yet)
```

```python
# Scatter plot: R&D Spend vs Profit
plt.scatter(data["R&D Spend"], data["Profit"])   # Plot R&D Spend on x-axis and Profit on y-axis
plt.title("Relationship between R&D Spend and Profit", fontsize=14)   # Set chart title
plt.xlabel("R&D Spend ($)", fontsize=12)         # Label x-axis
plt.ylabel("Profit ($)", fontsize=12)            # Label y-axis
plt.grid(True, alpha=0.3)                        # Add light transparent grid
plt.show()                                       # Display the plot
```

```python
# Scatter plot: Administration Spend vs Profit
plt.scatter(data["Administration"], data["Profit"])   # Plot Administration vs Profit
plt.title("Relationship between Administration and Profit", fontsize=14)
plt.xlabel("Administration", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.grid(True, alpha=0.3)                            # Light grid
plt.show()
```

```python
# Calculate the average profit for each state
state_profit = data.groupby("State")["Profit"].mean()   # Group by state and compute mean profit
state_profit   
```

```python
# Bar chart: Average profit by state
plt.bar(state_profit.index, state_profit.values, 
        color=['skyblue', 'orange', 'lightgreen'])      # Create bar plot with custom colors
plt.title("Average Profit by State", fontsize=16)
plt.xlabel("State", fontsize=12)
plt.ylabel("Average Profit ($)", fontsize=12)
plt.show()
```

```python
# Boxplot comparing spending categories
plt.boxplot([data["R&D Spend"], 
             data["Administration"],
             data["Marketing Spend"]],
            labels=["R&D Spend", "Administration", "Marketing Spend"],  # Specify labels
            patch_artist=True,                                         # Enable color fill
            boxprops=dict(facecolor="lightblue", color="#3498db"))      # Style for box color

plt.title("Comparison of Spending Categories", fontsize=16)
plt.ylabel("Amount Spent ($)", fontsize=12)
plt.grid(axis='y', alpha=0.3)   # Add horizontal grid lines only
plt.show()
```
